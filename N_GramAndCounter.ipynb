## Elegant n-gram generation in Python(http://www.locallyoptimal.com/blog/2013/01/20/elegant-n-gram-generation-in-python/)

```python
def generate_unigrams(x):
    n_grams = list(zip(*[x[i:] for i in range(1)]))
   
    return n_grams
```

```python
def generate_bigrams(x):
    n_grams = list(zip(*[x[i:] for i in range(2)]))
   
    return n_grams
```

```python
def generate_trigrams(x):
    n_grams = list(zip(*[x[i:] for i in range(3)]))
   
    return n_grams
```

```python
def generate_ngrams(input_list, n):
    return list(zip(*[input_list[i:] for i in range(n)]))
```

```python
def generate_ngrams_uniq(input_list, n):
    return set(zip(*[input_list[i:] for i in range(n)]))
```
